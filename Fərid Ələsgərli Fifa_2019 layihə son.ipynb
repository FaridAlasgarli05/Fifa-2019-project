## Preprocessing (Oyunçu dəyərinin rəqəmə çevrilməsi)

##### Value sutununa baxmaq isteyirem. Sutunun ilk 10 reqemini nece gore bilerem?

```python
df['Value'].head(10)
```

##### Burda Euro isaresini cixaririq, M herfini 1000000-la, K herfini ise 1000-le evezleyirik

```python
df['Value'] = df['Value'].str.replace("€","")
df['Value'] = df['Value'].str.replace("M"," 1000000")
df['Value'] = df['Value'].str.replace("K"," 1000")
df['Value']
```

##### İndi isə sütunu iki hissəyə böləcək kod yazın və bu iki sütunu bir birinə vuraraq real "Value" sütununu əldə edin (float tipində)

```python
df[['Value 1','Value 2']] = df['Value'].str.split(' ',expand = True)
df['Value 1'] = df['Value 1'].astype('float')
df['Value 2'] = df['Value 2'].astype('float')
df['Value'] = df['Value 1']*df['Value 2']
df.drop(['Value 1','Value 2'],axis=1,inplace = True)

```

```python
df.columns
```

##### Menim sutunumda null deyerler ola biler. Men isteyirem ki bunu 0 ile evez edim

```python
df.isnull().sum()
df['Value'].isnull().sum()  # Value sütununda 252 null dəyər var.

```

```python
#df['Value'].fillna(value=0,inplace = True)
df['Value'] = df['Value'].fillna(value=0)
```

```python
df['Value'].isnull().sum()  # Value sütunundakı bütün null dəyərləri 0 ilə əvəz etdik.
```
